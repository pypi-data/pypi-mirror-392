# Smoketest Report

This notebook collects the results of the individual smoketest tasks and presents them in a combined report.

```python
import os
from pathlib import Path
```

```python
pipeline_input_dir = Path(os.environ["INPUT_PIPELINE_DIR"])
year = Path(os.environ["DATA_YEAR"])
```

```python
task_input_dir = pipeline_input_dir / "smoketest"
```

```python
db_file = task_input_dir / "db.txt"
db_records = int(db_file.read_text())
```

```python
universe_file = task_input_dir / "universe.txt"
universe_records = int(universe_file.read_text())
```

```python
print(f"{db_records} incidents found in the database for {year}")
print(f"{universe_records} records found in the universe file for {year}")
```
