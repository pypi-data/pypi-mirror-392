```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "/data/isa/mlex/experiments/pcpe4/pcpe_4.csv"
df = pd.read_csv(path, delimiter=';')
df
```

```python
tipologias = ["I-d", "I-e", "IV-n"]

# Dicionário para armazenar resultados
resumo = {
    "Contas": [],
    "Individuos/empresas": [],
    "Transacoes": []
}

colunas = []

for tip in tipologias:
    df_tip = df[df[tip] == 1]
    
    contas = df_tip["NUMERO_CONTA"].nunique()
    individuos = df_tip["CPF_CNPJ_TITULAR"].nunique()
    transacoes = len(df_tip)
    
    resumo["Contas"].append(contas)
    resumo["Individuos/empresas"].append(individuos)
    resumo["Transacoes"].append(transacoes)
    colunas.append(tip)

# "Nenhuma"
df_nenhuma = df[(df["I-d"] == 0) & (df["I-e"] == 0) & (df["IV-n"] == 0)]
resumo["Contas"].append(df_nenhuma["NUMERO_CONTA"].nunique())
resumo["Individuos/empresas"].append(df_nenhuma["CPF_CNPJ_TITULAR"].nunique())
resumo["Transacoes"].append(len(df_nenhuma))
colunas.append("Nenhuma")

# Monta a tabela final
tabela = pd.DataFrame(resumo, index=colunas).T

print(tabela)
```
