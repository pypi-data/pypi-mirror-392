```python
import subprocess
```

```python
shell = subprocess.Popen(
    ['/bin/bash'],
    stdin=subprocess.PIPE,
    stdout=subprocess.PIPE,
    stderr=subprocess.PIPE,
    text=True,
    bufsize=1
)
```

```python
testling_command: str = "ls -la"
marker = "__END_OF_COMMAND__"

# send command to the shell
shell.stdin.write(testling_command + f"\necho {marker}\n")
shell.stdin.flush()

# read and print the output (hacky but simple)
output_lines: str = ""
try:
    max_counter = 0
    while True:
        max_counter += 1
        line = shell.stdout.readline()
        if not line or line == "" or marker in line:
            break
        output_lines += line
        if shell.poll() is not None:
            break
except KeyboardInterrupt:
    pass

print(output_lines)
```
