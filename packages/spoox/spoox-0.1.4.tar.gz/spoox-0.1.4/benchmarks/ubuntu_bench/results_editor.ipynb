```python
import pickle
from pathlib import Path
from pprint import pprint
import json
```

```python
current_dir = Path.cwd()
current_dir
```

```python
pickle_file = current_dir / f"results_20250825-124005_singleton_gpt-oss-20b_v-False_f-0_d-False" / "ubuntu_bench_run_meta_data.pkl"
```

```python
# load and print pickle file
with pickle_file.open("rb") as f:
    meta_data = pickle.load(f)
    pprint(meta_data)
```

```python
# modify dict
meta_data['runs'] = 3
meta_data['task_ids'] = range(1, 21)
```

```python
# modify dict
with pickle_file.open("wb") as f:
    pickle.dump(meta_data, f)
```

```python
# change prefix of all files in a dir
work_dir = Path.cwd() / "results_20250825-130009_singleton_gpt-oss-20b_v-False_f-0_d-False"
work_dir
```

```python
old_prefix = "run_1"
new_prefix = "run_2"
for f in work_dir.iterdir():
    if f.is_file() and f.name.startswith(old_prefix):
        new_name = new_prefix + f.name[len(old_prefix):]
        f.rename(f.with_name(new_name))
        print(f"Renamed: {f.name} -> {new_name}")
```

# --------

```python
# convert pickl file to json file
work_dir = Path.cwd() / "results_singleton_gpt-oss-20b_v-False_f-0_d-False"

suffix = "exec_meta_data.pkl"

for f in work_dir.iterdir():
    if f.is_file() and f.name.endswith(suffix):
        with f.open("rb") as fp:
            loaded_data = pickle.load(fp)
        with open(f.name[:-4] + ".json", "w") as fj:
            json.dump(loaded_data, fj, indent=4)
        f.unlink()
        
```
