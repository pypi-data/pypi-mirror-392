```python
import os
import pandas as pd
import tempfile
from flync.features.feature_wrapper import FeatureWrapper as fw
import flync.features.feature_cleaning as ct

# Create a temporary directory and return its path
# Use mkdtemp() so tmp_dir is a string path usable across notebook cells
tmp_dir = tempfile.mkdtemp()
```

```python
fw = fw(
    cache_dir=tmp_dir,
    threads=20
)
```

```python
ncr = fw.run_all(
    gtf_file='/home/chlab/flync/src/data/ncr_base.gtf',
    config_file='/home/chlab/flync/src/config/bwq_config.yaml',
    ref_genome_path='/home/chlab/flync/genome/Drosophila_melanogaster.BDGP6.46.dna.toplevel.fa',
    sparse=False,
    use_dim_redux=True,
    redux_n_components=1,
    use_tfidf=True
)
```

```python
pcg = fw.run_all(
    gtf_file='/home/chlab/flync/src/data/pcg_base.gtf',
    config_file='/home/chlab/flync/src/config/bwq_config.yaml',
    ref_genome_path='/home/chlab/flync/genome/Drosophila_melanogaster.BDGP6.46.dna.toplevel.fa',
    sparse=False,
    use_dim_redux=True,
    redux_n_components=1,
    use_tfidf=True
)
```

```python
plc = ct.PipelineConfig(
    positive_df=ncr,
    negative_df=pcg,
    label_column='y',
    train_size=0.7,
    test_size=0.15,
    val_size=0.15,
    split_output_dir='/home/chlab/flync/data',
    cleaned_output_path='/home/chlab/flync/data/cleaned_training_data.parquet',
    random_seed=99,
    scaler='standard',
    scaler_path='/home/chlab/flync/models/scaler.pkl',
    enable_multi_hot=True,
    mode='training',
)

res = ct.prepare_training_data(plc)
```
