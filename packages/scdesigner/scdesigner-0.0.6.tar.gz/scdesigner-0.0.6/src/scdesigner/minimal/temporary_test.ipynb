```python
from scdesigner.minimal.copula import Copula
from scdesigner.minimal.standard_copula import StandardCopula
import numpy as np
import scanpy as sc
adata = sc.datasets.pbmc3k()[:, :300]

model = StandardCopula()
model.setup_data(adata, {"group": "~ 1"})
model.fit(lambda y, _: np.random.uniform(size=y.shape), top_k=5)
```

```python
import torch
model.pseudo_obs({"group": torch.ones((10, 1))})
```

```python
class DummyCopula(Copula):
    def fit(self):
        pass
    def likelihood(self):
        pass
    def num_params(self):
        return 0
    def pseudo_obs(self, x_dict):
        return np.random.uniform(size=(x_dict["group"].shape[0], self.n_outcomes))

model = DummyCopula({"group": "~ 1"})
model.setup_data(adata, {"group": "~ 1"})
model.fit()
model.pseudo_obs({"group": torch.ones((10, 1))})
```
