```python
from lyrics.client import BashClient

async def main():
    # Connect to a running Lyrics server
    async with BashClient("http://localhost:8870") as client:
        # Execute a command
        result = await client.execute_bash("echo 'Hello from Lyrics!'")
        print(f"Output: {result.stdout}")

        # List available skills
        skills = await client.list_skills()
        print(f"Available skills: {[skill.name for skill in skills]}")
        for skill in skills:
            print(f"Skill: {skill.name}, Description: {skill.description}")

        # Execute a workflow of commands
        commands = [
            # "pwd",
            # "ls -la",
            # "mkdir -p test_dir",
            # "ls test_dir",
            "cd /var/",
            "ls -la"
        ]
        results = await client.execute_workflow(commands)
        print(f"Workflow completed with {len(results)} steps")
        print("Workflow Outputs:")
        for i, result in enumerate(results):
            print(f"Step {i+1}:")
            print(f"  Command: {commands[i]}")
            print(f"  Exit code: {result.exit_code}")
            print(f"  Output: {result.stdout}")
            print(f"  Error: {result.stderr}")

await main()
```
